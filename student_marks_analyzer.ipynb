```python
import numpy as np
```

* Student Marks Matrix
* Rows = Students
* Columns = Subjects

```python
marks = np.array([ [78, 85, 90],  [45, 56, 60], [92, 88, 95], [35, 40, 38],[67, 72, 70]])
```

```python
print("Student Marks Matrix:")
print(marks)
```

```python
## Shape of Matrix
print("Shape of matrix:")
marks.shape
```

```python
# Average Marks Per Student
# axis=1 
# means row-wise
```

```python
student_avg = np.mean(marks, axis=1)

print("Average marks of each student:")
print(student_avg)
```

```python
# Subject-wise Average
# axis=0 
#column-wise
```

```python
subject_avg = np.mean(marks, axis=0)

print("Subject-wise average:")
print(subject_avg)
```

```python
# Highest Marks

highest_marks = np.max(marks)

print("Highest marks in matrix:")
print(highest_marks)
```

```python
# Topper
topper_index = np.argmax(student_avg)

print("Topper Student Index:")
print(topper_index)

print("Topper Average:")
print(student_avg[topper_index])
```

```python
# Failed Students
# Condition: Any subject < 40
```

```python
failed_students = np.any(marks < 40, axis=1)
```

```python
print("Failed Students:")
print(failed_students)
```

```python
# Print failed student indexes
print("Indexes of failed students:")

for index in range(len(failed_students)):
    if failed_students[index]:
        print(index)
```

**Grade System**

```python
print("Student Grades:")

for i in range(len(student_avg)):

    avg = student_avg[i]

    if avg >= 90:
        grade = "A"

    elif avg >= 75:
        grade = "B"

    elif avg >= 50:
        grade = "C"

    else:
        grade = "Fail"

    print(f"Student {i} --> Average: {avg:.2f} --> Grade: {grade}")
```
